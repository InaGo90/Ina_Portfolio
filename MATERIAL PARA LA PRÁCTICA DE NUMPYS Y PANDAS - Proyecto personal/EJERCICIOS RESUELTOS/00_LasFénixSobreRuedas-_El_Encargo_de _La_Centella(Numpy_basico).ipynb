### ➕➖✖️➗ Parte 3: Operando con Arrays

NumPy destaca por su capacidad para realizar operaciones elemento a elemento de forma eficiente.

**Tareas:**
1.  Crea dos arrays 1D: `array_a = np.array([1, 2, 3, 4])` y `array_b = np.array([10, 20, 30, 40])`.
2.  Calcula la suma de `array_a` y `array_b` y guárdala en `suma_arrays`. Muestra el resultado.
3.  Calcula la resta de `array_b` menos `array_a` (`resta_arrays`). Muestra el resultado.
4.  Calcula el producto elemento a elemento de `array_a` y `array_b` (`producto_arrays`). Muestra el resultado.
5.  Multiplica `array_a` por un escalar (un solo número), por ejemplo, por 5 (`producto_escalar`). Muestra el resultado.
6.  Suma un escalar, por ejemplo 100, a `array_b` (`suma_escalar`). Muestra el resultado.

```python
#1

array_a = np.array([1, 2, 3, 4])
array_b = np.array([10, 20, 30, 40])
```

```python
#2

suma_arrays = array_a + array_b
print(suma_arrays)
```

```python
#3

resta_arrays = array_b - array_a
print(resta_arrays)
```

```python
#4

producto_arrays = array_a * array_b
print(producto_arrays)
```

```python
#5

producto_escalar = array_a * 5
print(producto_escalar)
```

```python
#6

suma_escalar = array_b + 100
print(suma_escalar)
```
